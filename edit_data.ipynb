```python
import pickle as pkl
from copy import deepcopy
```

```python

def display_example(x):
    display(x.load_seg_image())
    print(x.captions)
    print(repr(x.seg_info))
```

```python
# threeballs
```

```python
with open("evaldata/threeballs1.pkl", "rb") as f:
    data = pkl.load(f)
```

```python
for example in data:
    display_example(example)
```

```python
prefixes = ["a highly detailed photorealistic image of "]
# prefixes += ["a photo of standing on ", "a photo of floating in ", "a photo of in "]    # for openair1.pkl
prefixes += ["a photo of "]
prefixes += ["a digital art of "]

newdata = []
for example in data:
    example = deepcopy(example)
#     print(example)
    for rgbcode, seginfo in example.seg_info.items():
        caption = seginfo["caption"]
        captionchanged = 0
        for prefix in prefixes:
#             print(prefix, caption)
            if caption.startswith(prefix):
                seginfo["caption"] = seginfo["caption"][len(prefix):]
                captionchanged += 1
                break
        assert captionchanged == 1
    newdata.append(example)

for example in newdata:
    display_example(example)
    
```

```python
with open("evaldata/threeballs1b.pkl", "wb") as f:
    pkl.dump(newdata, f)
```
